# Basic List Operations 

1. Create a list of 5 numbers and add a new number to the end.


```python
lst=[1,2,3,4,5]
lst.append(9)
lst
```

2. Make a list of fruits and insert a fruit at the second position.


```python
lst=["apple","guava","mango","orange"]
lst.insert(1,"grapes")
lst
```

3. Create a list of names and remove one name from the list.


```python
lst=["sumedh","hemant","sourabh","kedar","daaku"]
print(lst)
lst.remove("daaku")
lst
```

4. Create a list of 4 items and change the value at index 2.


```python
items=['aaa','bbb','ccc','ddd']
print(items)
items[2]='xyz'
items
```

9. Create a list of animals and add three more animals using extend().


```python
animals1=['lion','tiger','elephant']
print(animals1)
animals2=['leopard','bison','cheetah']
print(animals2)
animals2.extend(animals1)
print(animals2)
```
