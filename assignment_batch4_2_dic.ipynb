6. Create a dictionary and remove the last inserted item using popitem().


```python
dic={i:i*i for i in range (1,9)}
print(dic)
dic.popitem()
dic
```

9. Create a dictionary and print all the key-value items.


```python
dic={"name":"John","age":24,"city":"pune"}
for key,val in dic.items():
    print(key,val)
```
